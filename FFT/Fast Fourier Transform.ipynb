```python
import pandas as pd
from numpy import sin, linspace, pi
from pylab import plot, show, title, xlabel, ylabel, subplot
from scipy import fft, arange

%matplotlib inline
```

```python
def plotSpectrum(y,Fs):
    """
    Plots a Single-Sided Amplitude Spectrum of y(t)
    source: http://glowingpython.blogspot.com/2011/08/how-to-plot-frequency-spectrum-with.html
    
    Fs = sampling rate
    """
    n = len(y) # length of the signal
    k = arange(n)
    T = n/Fs
    frq = k/T # two sides frequency range
    frq = frq[range(n/2)] # one side frequency range

    Y = fft(y)/n # fft computing and normalization
    Y = Y[range(n/2)]

    plot(frq,abs(Y),'r') # plotting the spectrum
    xlabel('Freq (Hz)')
    ylabel('|Y(freq)|')
```

```python
Fs = 150.0;  # sampling rate
Ts = 1.0/Fs; # sampling interval
t = arange(0,1,Ts) # time vector

ff = 5;   # frequency of the signal
y = sin(2*pi*ff*t)
```

```python
subplot(2,1,1)
plot(t,y)
xlabel('Time')
ylabel('Amplitude')
subplot(2,1,2)
plotSpectrum(y,Fs)
show()
```

```python
def fft_df(y,Fs):
    """
    Returns a pandas DataFrame w/ frq and Y.
    Make sure to import pandas as pd
    y = amp
    Fs = sampling frequency
    """
    n = len(y) # length of the signal
    k = arange(n)
    T = n/Fs
    frq = k/T # two sides frequency range
    frq = frq[range(n/2)] # one side frequency range

    Y = fft(y)/n # fft computing and normalization
    Y = Y[range(n/2)]
    Y = abs(Y)
    
    Y_fft = pd.DataFrame()
    Y_fft['frq'] = frq
    Y_fft['Y'] = Y
    
    return Y_fft
```

```python
def mc_frq(y,Fs):
    """
    Returns most common frq.
    Uses fft_df()
    y = amp
    Fs = sampling frequency
    """
    Y_df = fft_df(y,Fs)
    print 'Most common frequency is:'
    
    return Y_df.get_value(index=Y_df['Y'].idxmax(),col='frq')
```
