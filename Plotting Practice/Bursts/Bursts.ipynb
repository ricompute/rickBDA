```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
bursts = pd.read_csv('rat34_ECG.txt_Bursts_Results.csv')
bursts.rename(columns={'Unnamed: 1':'Time'}, inplace=True) #label the time column
bursts
```

```python
peak_amp = bursts['Peak Amp']
peak_time = bursts['Peak Time']

#Create a plot of the peak amps
fig1 = plt.figure()
ax1 = fig1.add_subplot(1,1,1)
plt.plot(peak_time,peak_amp);
ax1.set_xlim([0,610]), ax1.set_ylim([0.35,0.65]);
ax1.set_xlabel('Time'), ax1.set_ylabel('Peak Amp');
ax1.set_title('Rat 34 ECG Bursts');
```

```python
#Create a Peak Interval column, similar to Intraburst Interval.

bursts['Peak Interval'] = bursts['Peak Time']

for x in np.arange(0,len(bursts.index)-1):
    bursts.loc[x,'Peak Interval'] = bursts.loc[x+1,'Peak Time'] - bursts.loc[x,'Peak Time']

bursts.loc[len(bursts.index)-1,'Peak Interval'] = 'NaN'

bursts
```

```python
#Attempt a Poincare plot using peak time

fig2 = plt.figure()
ax2 = fig2.add_subplot(1,1,1)

lpt = len(peak_time)

for x in np.arange(0,lpt-3):
    plt.scatter((peak_time[x+1]-peak_time[x]),(peak_time[x+2]-peak_time[x+1]),edgecolors='r',facecolors='none');

plt.scatter((peak_time[lpt-1]-peak_time[lpt-2]),(peak_time[1]-peak_time[0]),edgecolors='r',facecolors='none');

ax2.set_title('Poincare Plot Attempt');
```
