```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
pop = pd.read_table('population_clean.txt', sep='\s+', names=['Pop Rank','Country','Population'], index_col='Pop Rank', thousands=',')
pop
```

```python
pop['Population'].dtype
```

```python
le = pd.read_table('life_expectancy_clean.txt', sep='\s+', names=['LE Rank','Country','Life Expectancy'], index_col='LE Rank', thousands=',')
le
```

```python
le['Life Expectancy'].dtype
```

```python
ed = pd.read_table('edu_expend_clean.txt', sep='\s+', names=['Ed Rank','Country','Ed Expenditures'], index_col='Ed Rank', thousands=',')
ed
```

```python
ed['Ed Expenditures'].dtype
```

```python
gdp_cap = pd.read_table('gdp_per_cap_clean.txt', sep='\s+', names=['GDP/cap Rank','Country','GDP/cap'], index_col='GDP/cap Rank', thousands=',')
gdp_cap
```

```python
gdp_cap = gdp_cap.dropna()
gdp_cap
```

```python
gdp_cap['GDP/cap'].dtype
```

```python
data = pd.merge(pd.merge(pd.merge(pop,le),ed),gdp_cap)
data
```

```python
data.corr()
```

```python
# plot histograms of GDP/cap and life expectancy

fig1 = plt.figure()

ax1 = fig1.add_subplot(2,1,1)
ax1.hist(data['GDP/cap'],bins=20);
ax1.set_xlabel('GDP per capita ($)'), ax1.set_ylabel('# Countries');

ax2 = fig1.add_subplot(2,1,2)
ax2.hist(data['Life Expectancy']);
ax2.set_xlabel('Life Expectancy (yr)'), ax2.set_ylabel('# Countries');

plt.subplots_adjust(wspace=0,hspace=0.5);
```

```python
# natural log transformation of GDP/cap
data['ln(GDP)'] = np.log(data['GDP/cap'])
data
```

```python
data.corr()
```

```python
# plot GDP/cap vs. life expectancy

fig2 = plt.figure()

ax3 = fig2.add_subplot(1,1,1)
ax3.scatter(data['GDP/cap'],data['Life Expectancy'], edgecolors='g', facecolors='none');
ax3.set_xlim(0,100000);
ax3.set_xlabel('GDP per capita ($)'), ax3.set_ylabel('Life Expectancy (yr)');
```

```python
# plot ln(GDP/cap) vs. life expectancy

fig3 = plt.figure()

ax4 = fig3.add_subplot(1,1,1)
ax4.scatter(data['ln(GDP)'],data['Life Expectancy'], edgecolors='g', facecolors='none');
ax4.set_xlabel('ln(GDP per capita)'); ax4.set_ylabel('Life Expectancy (yr)');
```
